```python
order = [i for i in df7.columns if df7[i].dtype == 'object']
new_order = [j for j in df7[order].values]
```

```python
new_order
```

```python
# Looping through all categorical column and convert it to numeric using Ordinal Encoding
for j in order:
  order_1  = df7[j].unique()
  ordinal = OrdinalEncoder(categories=[order_1])
  encoded = ordinal.fit_transform(df7[[j]])
  df7[j] = encoded
```
